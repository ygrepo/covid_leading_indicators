```python
filename = "data/us-counties.csv"
covid_df = pd.read_csv(filename)
```

```python
def process_data(state:str, postal_state:str):
    state_policy_df = policy_df[policy_df["StateName"] == state]
    state_policy_df.drop(["location_id", "StateFIPS", 
                        "StatePostal", "PolicyCodingNotes","PolicySource","StateWide","LastUpdated","LastUpdatedNotes"], axis=1, inplace=True)
    state_policy_df["DateIssued"] =  pd.to_datetime(state_policy_df["DateIssued"], format='%Y%m%d')
    state_policy_df["DateEnacted"] =  pd.to_datetime(state_policy_df["DateEnacted"], format='%Y%m%d')
    state_policy_df["DateExpiry"] =  pd.to_datetime(state_policy_df["DateExpiry"], format='%Y%m%d')
    state_policy_df["DateEased"] =  pd.to_datetime(state_policy_df["DateEased"], format='%Y%m%d')
    state_policy_df["DateEnded"] =  pd.to_datetime(state_policy_df["DateEnded"], format='%Y%m%d')
    end = pd.to_datetime("now")
    end = end.to_datetime64()
    state_policy_df['DateExpiry'] = state_policy_df['DateExpiry'].fillna(end)
    state_policy_df['DateExpiry'] = pd.to_datetime(state_policy_df['DateExpiry'])
    state_policy_df['DateEased'] = state_policy_df['DateEased'].fillna(end)
    state_policy_df['DateEased'] = pd.to_datetime(state_policy_df['DateEased'])
    state_policy_df['DateEnded'] = state_policy_df['DateEnded'].fillna(end)
    state_policy_df['DateEnded'] = pd.to_datetime(state_policy_df['DateEnded'])
    state_policy_df['DateExpiry'] = pd.to_datetime(state_policy_df['DateExpiry'])
    state_rt_df = rt_df[rt_df["region"] == postal_state]
    state_rt_df["date"] = state_rt_df["date"].astype("datetime64")
    state_rt_df = state_rt_df.set_index("date")
    state_rt_df.sort_index(inplace=True)
    state_covid_df = covid_df[covid_df["state"] == state]
    state_covid_df["date"] = state_covid_df["date"].astype("datetime64")
    state_covid_df = state_covid_df.set_index("date")
    state_covid_df.sort_index(inplace=True)
    state_covid_df = state_covid_df.groupby(pd.Grouper(freq='D')).sum()
    state_covid_df["cases_pct_changes"] = state_covid_df["cases"].pct_change()
    state_covid_df["deaths_pct_changes"] = state_covid_df["deaths"].pct_change()
    state_covid_df["case_change"] = state_covid_df["cases"].diff()
    state_covid_df["deaths_change"] = state_covid_df["deaths"].diff()
    state_covid_df.drop(["fips"], axis=1, inplace=True)
    return (state_policy_df, state_rt_df, state_covid_df)
```

```python
state_policy_df, state_rt_df, state_covid_df = process_data("New York", "NC")
```

```python
def create_policy_chart(state:str, postal_state:str):
    state_policy_df, state_rt_df, state_covid_df = process_data(state, postal_state)
    policies_chart = alt.Chart(state_policy_df).mark_bar(size=2).encode(
        alt.X("DateIssued:T", title="DATE", axis=alt.Axis(title='Date', orient="top")),
        x2='DateExpiry',
        y='StatePolicy'
    )
    state_rt_chart = alt.Chart(state_rt_df.reset_index()).mark_line(strokeWidth=4).encode(
        alt.X("date:T"), 
        alt.Y("new_cases:Q", title="DAILY NEW CASES"), 
        color=alt.value('red')
    )
    alt_layer = alt.layer(
      state_rt_chart,
      policies_chart,
    ).properties(
        width=400,
        height=100,
        title=state
    )
    return alt_layer
```

```python
state_dict={'Alabama':"AL",
 'Alaska':"AK",
 'Arizona':"AZ",
 'Arkansas':"AR",
 'California':"CA",
 'Colorado':"CO",
 'Connecticut':"CT",
 'Delaware':"DE",
 'District of Columbia':"DC",           }
```

```python
def create_compound_charts(base_chart_fn:Callable):
    # make a single row
    def make_hcc(row_of_charts):
        hconcat = [chart for chart in row_of_charts]
        hcc = alt.HConcatChart(hconcat=hconcat)
        return hcc
    # take an array of charts and produce a facet grid
    def facet_wrap(charts, charts_per_row):
        rows_of_charts = [
            charts[i:i+charts_per_row] 
            for i in range(0, len(charts), charts_per_row)]        
        vconcat = [make_hcc(r) for r in rows_of_charts]    
        vcc = alt.VConcatChart(vconcat=vconcat)\
          .configure_axisX(grid=True)\
          .configure_axisY(grid=True)
        return vcc

    #create a list of subplots
    charts = []
    states = state_dict.keys()
    postal_states = state_dict.values()
    for state, postal_state in zip(states, postal_states):
        charts.append(base_chart_fn(state, postal_state))


    compound_chart = facet_wrap(charts, charts_per_row=2)
    compound_chart.properties(title='DAILY NEW CASES - STATE POLICIES')
    return compound_chart
```

```python
def create_policy_rt_chart(state:str, postal_state:str):
    state_policy_df, state_rt_df, state_covid_df = process_data(state, postal_state)
    policies_chart = alt.Chart(state_policy_df).mark_bar(size=2).encode(
        alt.X("DateIssued:T", title="DATE", axis=alt.Axis(title='Date', orient="top")),
        x2='DateExpiry',
        y='StatePolicy'
    )
    state_rt_chart = alt.Chart(state_rt_df.reset_index()).mark_line(strokeWidth=4).encode(
    alt.X("date:T"), 
    alt.Y("mean:Q", title="Mean Rt Covid-19"), 
    color=alt.value('red')
    )
    alt_layer = alt.layer(
      state_rt_chart,
      policies_chart,
    ).properties(
        width=400,
        height=100,
        title=state
    )
    return alt_layer
```

```python
def create_policy_covid_chart(state:str, postal_state:str):
    state_policy_df, state_rt_df, state_covid_df = process_data(state, postal_state)
    policies_chart = alt.Chart(state_policy_df).mark_bar(size=2).encode(
        alt.X("DateIssued:T", title="DATE", axis=alt.Axis(title='Date', orient="top")),
        x2='DateExpiry',
        y='StatePolicy'
    )
    covid_chart = alt.Chart(state_covid_df.reset_index()).mark_line(strokeWidth=4).transform_fold(
        fold=['cases_pct_changes', 'deaths_pct_changes'], 
        as_=['variable', 'value']
    ).encode(
        alt.X("date:T", title="DATE", axis=alt.Axis(title='Date', orient="top")),
        alt.Y("value:Q", title="Pct Changes"), 
        color='variable:N'
    )
    alt_layer = alt.layer(
      covid_chart,
      policies_chart,
    ).properties(
        width=400,
        height=100,
        title=state
    )
    return alt_layer
```

```python
compound_chart = create_compound_charts(create_policy_covid_chart)
compound_chart
```
